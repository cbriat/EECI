# AIC Variance Heatmap (SSA)



**Figures:** aic_variance_heatmap.svg

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib, matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

matplotlib.rcParams.update({
    "svg.fonttype": "path", "mathtext.fontset": "cm",
    "font.family": "serif", "font.size": 14,
    "axes.labelsize": 16, "axes.titlesize": 15,
    "xtick.labelsize": 13, "ytick.labelsize": 13,
    "axes.linewidth": 0.8,
})
OUT = "."

data = np.load(".")
k_vals, eta_vals, ratio = data["k_vals"], data["eta_vals"], data["ratio"]

fig, ax = plt.subplots(figsize=(9, 7))
fig.subplots_adjust(left=0.12, right=0.91, top=0.90, bottom=0.12)

im = ax.pcolormesh(k_vals, eta_vals, ratio,
                    cmap="YlOrRd",
                    norm=LogNorm(vmin=1.0, vmax=min(ratio.max(), 15)),
                    shading="gouraud")

# Contours
levels = [1.5, 2.0, 3.0, 5.0, 8.0]
levels = [l for l in levels if l <= ratio.max()*0.95]
CS = ax.contour(k_vals, eta_vals, ratio, levels=levels,
                colors=["0.15"], linewidths=[1.2], linestyles=":")
ax.clabel(CS, fmt="%.1f", fontsize=9)

# Find and mark the minimum
imin = np.unravel_index(np.argmin(ratio), ratio.shape)
k_best, eta_best = k_vals[imin[1]], eta_vals[imin[0]]
ax.plot(k_best, eta_best, "*", color="k", ms=14,
        markeredgecolor="white", markeredgewidth=1, zorder=5)
ax.text(k_best*1.3, eta_best*0.7,
        rf"min $= {ratio.min():.2f}$",
        fontsize=10, color="k",
        bbox=dict(fc="white", ec="0.5", pad=2,
                  boxstyle="round,pad=0.3"))

ax.set_xscale("log"); ax.set_yscale("log")
ax.set_xlabel(r"Actuation gain $k$")
ax.set_ylabel(r"Sequestration rate $\eta$")
ax.set_title(
    r"AIC noise: $\mathrm{Var}[P]_{\mathrm{CL}} / "
    r"\mathrm{Var}[P]_{\mathrm{OL}}$"
    r"  (SSA, ergodic for all $k, \eta > 0$)",
    fontsize=13, fontweight="bold")

cbar = fig.colorbar(im, ax=ax, extend="max", shrink=0.85)
cbar.set_label(r"$\mathrm{Var}_{\mathrm{CL}} / \mathrm{Var}_{\mathrm{OL}}$",
               fontsize=14)

fig.savefig(f"{OUT}/aic_variance_heatmap.svg", bbox_inches="tight")
fig.savefig(".", dpi=200,
            bbox_inches="tight")
plt.close()
print("Saved aic_variance_heatmap.svg")
```
